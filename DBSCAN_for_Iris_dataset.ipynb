```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
data=pd.read_csv("/content/Iris.csv")
data
```

```python
features = ['PetalLengthCm','PetalWidthCm']
x=data[features]
x_standardized=StandardScaler().fit_transform(x)
dbscan=DBSCAN(eps=0.5,min_samples=5)
clusters=dbscan.fit_predict(x_standardized)
clusters
```

```python
centroids = pd.DataFrame(x_standardized).groupby(clusters).mean().values
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x['PetalLengthCm'],x['PetalWidthCm'],c=clusters,cmap='viridis')
plt.title('DBSCAN Clustering of mall customer data')
plt.xlabel('PetalLengthCm)')
plt.ylabel('PetalWidthCm')
plt.show()
```
